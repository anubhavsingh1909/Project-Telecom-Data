```python
tc = sns.kdeplot(df.TotalCharges, color='green', legend = True)
kdeline = tc.lines[0]
mean = df.TotalCharges.mean()
median = df.TotalCharges.median()
xd = kdeline.get_xdata()
yd = kdeline.get_ydata()
h1 = np.interp(mean, xd, yd)
h2 = np.interp(median, xd, yd)
tc.vlines(mean, 0, h1, color='grey', label='mean')
tc.vlines(median, 0, h2, color='black', label='median')
tc.fill_between(xd, 0, yd, facecolor='green', alpha=0.1)
plt.legend()
plt.show()
```

```python
tc = sns.kdeplot(df.MonthlyCharges, color='orange', legend = True)
kdeline = tc.lines[0]
mean = df.MonthlyCharges.mean()
median = df.MonthlyCharges.median()
xd = kdeline.get_xdata()
yd = kdeline.get_ydata()
h1 = np.interp(mean, xd, yd)
h2 = np.interp(median, xd, yd)
tc.vlines(mean, 0, h1, color='grey', label='mean')
tc.vlines(median, 0, h2, color='black', label='median')
tc.fill_between(xd, 0, yd, facecolor='orange', alpha=0.1)
plt.legend()
plt.show()
```

```python
tc = sns.kdeplot(df.tenure, color='blue', legend = True)
kdeline = tc.lines[0]
mean = df.tenure.mean()
median = df.tenure.median()
xd = kdeline.get_xdata()
yd = kdeline.get_ydata()
h1 = np.interp(mean, xd, yd)
h2 = np.interp(median, xd, yd)
tc.vlines(mean, 0, h1, color='grey', label='mean')
tc.vlines(median, 0, h2, color='black', label='median')
tc.fill_between(xd, 0, yd, facecolor='blue', alpha=0.1)
plt.legend()
plt.show()
```
